GRÁFICO DA ANÁLISE

```python
# Filtra os 10 estados com maior número de mortes no último dia da base
top10 = df_latest.sort_values(by='deaths', ascending=False).head(10)

# Define o tamanho da figura (largura=10, altura=6)
plt.figure(figsize=(10, 6))

# Cria o gráfico de barras horizontais com os dados do top10
# x = número de mortes, y = nome do estado
# palette='Reds_r' define uma paleta de cores em tons de vermelho (reverso)
sns.barplot(x='deaths', y='state', data=top10, palette='Reds_r')

# Título do gráfico
plt.title('Top 10 Estados com Mais Mortes no Último Dia')

# Rótulo do eixo X
plt.xlabel('Número de Mortes')

# Rótulo do eixo Y
plt.ylabel('Estado')

# Ajusta o layout para evitar cortes de texto
plt.tight_layout()

# Mostra o gráfico
plt.show()
```
